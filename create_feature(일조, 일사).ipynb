```python
# -*- coding: utf-8 -*-

import os
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dangjin_df = pd.read_csv("data/energy_dangjin.csv")
floating_df = pd.read_csv("data/energy_floating.csv")
warehouse_df = pd.read_csv("data/energy_warehouse.csv")
```

```python
dangjin_df
```

```python
sunlight1 = pd.read_csv("data/seosan(dangjin) 18. 3. 1. - 19. 1. 31. 일조, 일사량 data.csv")
sunlight2 = pd.read_csv("data/seosan(dangjin) 19. 2. 1. - 20. 1. 31. 일조, 일사량 data.csv")
sunlight3 = pd.read_csv("data/seosan(dangjin) 20. 2. 1. - 21. 1. 31. 일조, 일사량 data.csv")
sunlight4 = pd.read_csv("data/seosan(dangjin) 21. 2. 1. - 21. 3. 1. 일조, 일사량 data.csv")
```

```python
sunlight = pd.concat([sunlight1, sunlight2, sunlight3, sunlight4], ignore_index=True)
```

```python
sunlight
```

```python
sunlight['date'] = pd.to_datetime(sunlight['일시'])
```

```python
sunlight_df = sunlight.drop(['일시'], axis=1)
```

```python
sunlight_df
```

```python
dangjin_sunlight_df = pd.concat([dangjin_df, sunlight_df], axis=1)
```

```python
dangjin_sunlight_df
```

```python
columns = ['date']
index = pd.date_range('2018-03-02', periods = 26304, freq = 'H',)
temp_df = pd.DataFrame(index=index, columns=columns)
temp_df = temp_df.reset_index(drop=True)

temp_df['date'] = pd.date_range('2018-03-02',
                     periods = 26304,
                     freq = 'H',)
```

```python
temp_df
```

```python
sunlight_df_ = temp_df.merge(sunlight_df, on='date', how='outer') # temp_df의 date열 값과 sunlight의 date열 값이 같으면 tempd_df의 date열에 일조, 일사량을 삽입함
```

```python
sunlight_df_
```

```python
sunlight_df_ = sunlight_df_.fillna(0)
```

```python
sunlight_df_
```

```python
sunlight_df_.drop(sunlight_df_.loc[sunlight_df_['date'] < '2018-03-02 00:00:00'].index, inplace=True)
```

```python
dangjin_sunlight_df = pd.concat([dangjin_df, sunlight_df_], axis=1)
floating_sunlight_df = pd.concat([floating_df, sunlight_df_], axis=1)
warehouse_sunlight_df = pd.concat([warehouse_df, sunlight_df_], axis=1)
```

```python
floating_sunlight_df
```

```python
dangjin_sunlight_df = dangjin_sunlight_df.drop(['date', '지점', '지점명'], axis=1)
floating_sunlight_df = floating_sunlight_df.drop(['date', '지점', '지점명'], axis=1)
warehouse_sunlight_df = warehouse_sunlight_df.drop(['date', '지점', '지점명'], axis=1)
```

```python
dangjin_sunlight_df.to_csv("data/energy_dangjin_sunlight.csv", index=False)
floating_sunlight_df.to_csv("data/energy_floating_sunlight.csv", index=False)
warehouse_sunlight_df.to_csv("data/energy_warehouse_sunlight.csv", index=False)
```
