```python
import os
import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
energy_floating = pd.read_csv('data/energy_floating_sunlight.csv') # 일조, 일사량 추가 된 데이터
energy_warehouse = pd.read_csv('data/energy_warehouse_sunlight.csv') # 일조, 일사량 추가 된 데이터
energy_dangjin = pd.read_csv('data/energy_dangjin_sunlight.csv') # 일조, 일사량 추가 된 데이터
energy_ulsan = pd.read_csv('data/energy_ulsan.csv')
```

```python
energy_floating_wind = energy_floating.copy()

wv = energy_floating_wind.pop('WindSpeed')

# Convert to radians.
wd_rad = energy_floating_wind.pop('WindDirection')*np.pi / 180

# Calculate the wind x and y components.
energy_floating_wind['Wind_X'] = wv*np.cos(wd_rad)
energy_floating_wind['Wind_Y'] = wv*np.sin(wd_rad)
```

```python
energy_floating_wind.to_csv("data/energy_floating_sunlight_wind.csv", index=False)
```
